# Load and display the CSV

Notes that matter in real workflows
* inferSchema is convenient for exploration but slower on large files; in production you’d define a schema explicitly.

* df.display() works in Databricks notebooks; outside Databricks use df.show() instead.

* If your file is in DBFS (e.g., uploaded to /FileStore), your path might look like:

``` python
"/dbfs/FileStore/customers-100.csv"
```

* If the file is in Workspace Files, use:

``` python
"file:/Workspace/Users/<your-user>/customers-100.csv"

```

````python
df = (
    spark.read
        .option("header", "true")        # CSV has column names
        .option("inferSchema", "true")   # Let Spark infer data types
        .csv("/Volumes/greentree/default/learnpysparks/customers-100.csv")
)

df.display()
````
